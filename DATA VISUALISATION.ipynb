```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.array([0,6])
y=np.array([0,30])
```

```python
plt.plot(x,y)
plt.show()
```

## Line plot

```python
arr=np.array([4,5,3,6,8,3,2,5,7])
```

```python
plt.plot(arr)
plt.show()
```

```python
plt.plot(arr,linestyle="dotted")
```

```python
plt.plot(arr,linestyle="dashed")
```

## Bar plot

```python
x=np.array(["M","T","G","R"])
y=np.array([4,6,3,8])
```

```python
plt.bar(x,y)
plt.show()
```

## Scatter Plot

```python
x=np.array([4,6,3,8,4,2,5,7])
y=np.array([8,4,2,2,4,5,2,2])
```

```python
plt.scatter(x,y)
plt.show()
```

## Pie Plot

```python
p=np.array([10,10,30,60])
plt.pie(p)
plt.show()
```

## Histogram

```python
h=np.random.normal(2,13,64)
plt.hist(h)
plt.show()
```
